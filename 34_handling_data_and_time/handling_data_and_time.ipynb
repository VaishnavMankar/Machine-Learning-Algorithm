## Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1. Extract year

```python
date['date_year'] = date['date'].dt.year
#creating new column of date_year which just store year in it
date.sample(5)
```

### 2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
#creating new column of month which store only month
date.head()
```

### Extract Days

```python
date['date_day'] = date['date'].dt.day
#just addted a column month which has month as a string 
date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()
# date_doen_new column tels us the wether its a first day of the weak or last day of weak 
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)
#This will give ous the day of the weak as a string 
date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week  # ✅ Correct way
# date['date_week'] = date['date'].dt.week

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
