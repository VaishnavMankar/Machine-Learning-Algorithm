```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Data Preparation
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3 * X[:, 0] ** 2 + 0.05 * np.random.randn(100)
```

```python
# Model
gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbr.fit(X, y)
```

```python
print(f"Number of decision trees used: {len(gbr.estimators_)}")
```

```python
# Predictions
X_new = np.linspace(-0.5, 0.5, 200).reshape(-1, 1)
y_pred = gbr.predict(X_new)
```

```python
# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Data', alpha=0.6)
plt.plot(X_new, y_pred, color='red', label='Gradient Boost Prediction', linewidth=2)
plt.title("Gradient Boosting Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Evaluation metrics
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R² Score: {r2:.4f}")
```
